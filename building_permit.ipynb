```python
'''Dropping such percentage of values from the dataset is not really adviceable, so we are going to drop columns
that have missing values that are more than half of the total row length'''

def drop_col(data):
    s = data.isnull().sum()
    cols = [i for i,j in s.items() if j>len(data)/2]
    data.drop(cols, axis=1, inplace=True)
```

```python
drop_col(permit)
```

```python
def freq_enc(df,cols):
    for col in cols:
        df[col] = df[col].map(df[col].value_counts().to_dict())/len(df)
    return df.head()
```

```python
cols = ['Permit Type Definition']
freq_enc(permit, cols)
```
